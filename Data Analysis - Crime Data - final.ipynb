```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
crime_data_to_load = "data/crime_data.csv"

crime_data = pd.read_csv(crime_data_to_load)
crime_data.head(500)
```

```python
#Pulling first four digits of INCIDENT NUMBER which indicates the year
pd.options.mode.chained_assignment = None
crime_data["INCIDENT_YEAR"] = crime_data["INCIDENT_NUMBER"].str[:4]
crime_data.head()
```

```python
grouped_crime_year = crime_data.groupby(["INCIDENT_YEAR"])
grouped_crime_year.count().head(19)
```

```python
incident_count = grouped_crime_year["INCIDENT_NUMBER"].count()
incident_count.head()
```

```python
incident_count_df = pd.DataFrame({"Number of Incidents": incident_count})
incident_count_df.head()
```

```python
years = ["2000", "2001", "2002", "2003", "2004", "2005", "2006",
         "2007", "2008", "2009", "2010", "2011", "2012", "2013",
         "2014", "2015", "2016", "2017", "2018"]
crime_count = incident_count_df.loc[years,["Number of Incidents"]]
```

```python
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),    
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),    
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),    
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),    
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]

for i in range(len(tableau20)):    
    r, g, b = tableau20[i]    
    tableau20[i] = (r / 255., g / 255., b / 255.)  

tableau20
```

```python
plt.plot(years, crime_count, color=tableau20[0], label="Crime Incidents")

plt.axis("tight")
plt.xticks(["2000","2002", "2004", "2006", "2008", "2010",
            "2012", "2014", "2016", "2018"])

plt.title("Scott's Addition Crime Incidents since 2000")
plt.xlabel("Years")
plt.ylabel("Number of Crime Incidents")
plt.text(0.02, 0.5, "Data source: http://www.richmondgov.com/Police/CrimeIncidentInfo.aspx"    
       "\nAuthor: City of Richmond", fontsize=10) 

plt.grid(True)
plt.savefig("..\\Scotts-Addition-RVA-Analysis\\plots\\scotts_addition_crime_incidents_sine_2000.jpeg")

plt.show()
```

```python
early_crime = crime_data.loc[crime_data["INCIDENT_YEAR"] <= "2009", :]
late_crime = crime_data.loc[crime_data["INCIDENT_YEAR"] > "2009", :]
```

```python
early_type_list = early_crime["OFFENSE_CODE_DESC"].value_counts()
early_top_five = early_type_list.head(5)
early_labels = early_top_five.index
explode = (0.2, 0, 0, 0, 0)

plt.pie(early_top_five, startangle=120, colors=tableau20, labels= early_labels, explode=explode, autopct="%1.1f%%",
        shadow=True, wedgeprops = {"linewidth": .5, "edgecolor": "black"})
plt.title("% of Top Crime Types from 2000 to 2009")
plt.savefig("..\\Scotts-Addition-RVA-Analysis\\plots\\percent_top_crime_types_2000_2009.jpeg")
plt.show()
```

```python
late_type_list = late_crime["OFFENSE_CODE_DESC"].value_counts()
late_top_five = late_type_list.head(5)
late_labels = late_top_five.index
explode = (0.2, 0, 0, 0, 0)

colors = [tableau20[0], tableau20[2], tableau20[7], tableau20[8], 
                    tableau20[3], tableau20[18]]

plt.pie(late_top_five, startangle=120, colors=colors, labels=late_labels, explode=explode, autopct="%1.1f%%",
        shadow=True, wedgeprops = {"linewidth": .5, "edgecolor": "black"})
plt.title("% of Top Crime Types from 2010 to 2018")
plt.savefig("..\\Scotts-Addition-RVA-Analysis\\plots\\percent_top_crime_types_2010_2018.jpeg")
plt.show()
```

```python
fig, ax = plt.subplots()

size = 0.3

ax.pie(late_top_five, radius=1, colors=colors, wedgeprops=dict(width=size, edgecolor='w'))

ax.pie(early_top_five, radius=1-size, colors=tableau20, wedgeprops=dict(width=size, edgecolor='w'))

ax.set(aspect="equal", title="2000-2009 and 2010-2018 Crime Types Compared")
plt.savefig("..\\Scotts-Addition-RVA-Analysis\\plots\\percent_top_crime_types_compare.jpeg")


plt.show()
```
